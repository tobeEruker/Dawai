```python
# 贴标签
print(features1.shape,features2.shape,features3.shape,features4.shape,features5.shape,features6.shape,features7.shape,features8.shape,features9.shape
     ,features10.shape,features11.shape,features12.shape,features13.shape,features14.shape,features15.shape)

X = np.concatenate((features1,features2,features3,features4,features5,features6,features7,features8,features9,features10,features11,features12,features13,features14,features15),axis=0)

labels = []
for class_idx in range(15):  # 类别索引 0~14
    labels.extend([class_idx] * 200)  # 生成标签列表，例如 [0,0,...,0, 1,1,...,1, ..., 14,14,...,14]
y = np.array(labels)                       # 形状: (总样本数,)
y = y.reshape(-1,1)
# 生成训练数据
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, num_classes=15)

print(y)
print(X.shape,y.shape)
```

```python
#分割训练集和测试集数据
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=5)
print(x_train.shape,x_test.shape,X.shape)
print(x_train.shape)
print(y_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense  #Conv2D卷积层 #MaxPooling2D池化层
model = Sequential()
model.add(Dense(units=128,activation='relu'))# units=128: 神经元的数量（128维）。# activation='relu': 使用 ReLU 激活函数。
model.add(Dense(units=15,activation='softmax'))
```

```python
#配置模型参数
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=15)
```

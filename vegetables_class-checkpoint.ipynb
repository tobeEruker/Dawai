```python
 from keras.preprocessing.image import ImageDataGenerator        #豆子苦瓜葫芦茄子西兰花卷心菜辣椒胡萝卜花椰菜黄瓜木瓜土豆南瓜萝卜和西红柿
                                                            # Sequential: Keras 的序列模型类，用于按顺序堆叠神经网络层。

                                                            # Conv2D: 二维卷积层，用于提取图像的空间特征。

                                                            # MaxPooling2D: 二维最大池化层，用于降维和减少计算量。

                                                            # Flatten: 将多维数据展平为一维，供全连接层使用。

                                                            # Dense: 全连接层，用于分类或回归。

# 图像增强、预处理（数值归一化，缩放，旋转，平移等）
train_datagen = ImageDataGenerator(rescale=1./255)      #归一化处理
# 加载图像
training_set = train_datagen.flow_from_directory('D:\\machine learn\\jupyter_precatice\\Vegetable Images\\train', #路径
                                                target_size = (50,50),   #图片尺寸
                                                batch_size = 32,         #选择多少图片进行梯度下降
                                                class_mode = 'categorical'           #根据目录结构自动标注类别                
)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense  #Conv2D卷积层 #MaxPooling2D池化层
# 搭建模型框架，之后根据结构再添
model = Sequential()
# 卷积层
model.add(Conv2D(32,(3,3),input_shape=(50,50,3),activation='relu')) 
                        #32: 卷积核（过滤器）的数量，决定输出特征图的深度。
                        # (3, 3): 每个卷积核的大小（高度×宽度）。
                        # input_shape=(50, 50, 3): 输入图像的形状（高度=50，宽度=50，通道数=3，即RGB图像）。
                        # activation='relu': 使用 ReLU 激活函数（max(0, x)），引入非线性。

# 池化层
model.add(MaxPooling2D(pool_size=(2,2)))  #pool_size=(2, 2): 池化窗口大小（高度×宽度），将每个 2×2 区域的最大值输出。
# 第二个卷积层、池化层
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
# Flattening层
model.add(Flatten())   #将多维特征图转换为一维向量。若上一层的输出是 (None, 12, 12, 32)，展平后变为 (None, 12*12*32 = 4608)。
# 全连接层
model.add(Dense(units=128,activation='relu'))# units=128: 神经元的数量（128维）。# activation='relu': 使用 ReLU 激活函数。
model.add(Dense(units=15,activation='softmax'))# activation='sigmoid': 使用 Sigmoid 函数，将输出压缩到 [0,1]，表示概率
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(training_set,epochs=20)
```

```python
#测试集下的准确率
test_set = train_datagen.flow_from_directory('D:\\machine learn\\jupyter_precatice\\Vegetable Images\\test', #路径
                                                target_size = (50,50),   #图片尺寸
                                                batch_size = 32,         #选择多少图片进行梯度下降
                                                class_mode = 'categorical'     #根据目录结构自动标注类别                       
)
accuracy_test=model.evaluate(test_set)
print(accuracy_test)
```
